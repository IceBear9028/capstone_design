```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('상관관계.csv', encoding = 'cp949')
```

```python
df.info()
```

```python
# corr 값에 따른 상관관계 출력 함수

def corr_discrimination(corr):
    abs_corr = abs(corr)
    if abs_corr > 0.2:
        if abs_corr > 0.4:
            if abs_corr > 0.6:
                if abs_corr > 0.85:
                    if abs_corr == 1:
                        return None
                    else:
                        return "매우강한상관관계"
                else:
                    return "높은상관관계"
            else:
                return "다소높은상관관계"
        else:
            return "낮은상관관계"
    else:
        return "매우낮은상관관계"
```

```python
df_discrimination = df.drop(['Unnamed: 0'], axis = 'columns')

for i in df_discrimination.columns:
    df_discrimination[i] = df_discrimination[i].apply(corr_discrimination)
```
